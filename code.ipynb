## Data preprocessing

```python
# seperate the datsa the label
X  = wine_dataset.drop('quality' , axis = 1)
```

```python
print(X)
```

## label binarization

```python
Y = wine_dataset['quality'].apply(lambda y_values:1 if y_values>=7 else 0)
```

```python
print(Y)
```

## train and test data

```python
X_train , X_test , Y_train , Y_test = train_test_split(X ,Y , test_size=0.2  , random_state=2)
```

```python
print(Y.shape ,  Y_train.shape, Y_test.shape)
```

## model testing
# random forestmodel

```python
model  = RandomForestClassifier()
```

```python
model.fit(X_train , Y_train)
```

## model evaluation

## acuracy score

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accurcy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('accuracy :', test_data_accurcy)
```

## Building a predicted system

```python
input_data = (7.5,0.5,0.36,6.1,0.071,17.0,102.0,0.9978,3.35,0.8,10.5)

# changing the input data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the data as we are predicting the label for only one instance
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

prediction = model.predict(input_data_reshaped)
print(prediction)

if (prediction[0]==1):
  print('Good Quality Wine')
else:
  print('Bad Quality Wine')
```
